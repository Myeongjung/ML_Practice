##**Boston**

```
import tensorflow as tf
import pandas as pd
```

```
file_location = 'https://raw.githubusercontent.com/Myeongjung/ML_Practice/master/csv/boston.csv'
boston = pd.read_csv(file_location)
print(boston.columns)
boston.head()
```

```
boston_indep = boston[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'b', 'lstat']]
boston_dep = boston[['medv']]

print(boston_indep.shape, boston_dep.shape)
```

##**Previous Structure**
X = tf.keras.layers.Input(shape=[13])

H = tf.keras.layers.Dense(10, activation='swish')(X)

Y = tf.keras.layers.Dense(1)(H)

```
X = tf.keras.layers.Input(shape=[13])

layer1 = tf.keras.layers.Dense(10)(X)
layer1 = tf.keras.layers.BatchNormalization()(layer1)
layer1 = tf.keras.layers.Activation('swish')(layer1)

layer2 = tf.keras.layers.Dense(10)(layer1)
layer2 = tf.keras.layers.BatchNormalization()(layer2)
layer2 = tf.keras.layers.Activation('swish')(layer2)

layer3 = tf.keras.layers.Dense(10)(layer2)
layer3 = tf.keras.layers.BatchNormalization()(layer3)
layer3 = tf.keras.layers.Activation('swish')(layer3)

Y = tf.keras.layers.Dense(1)(layer3)

boston_model = tf.keras.models.Model(X,Y)
boston_model.compile(loss='mse')#, metrics='accuracy')
```

```
boston_model.summary()
```

```
boston_model.fit(boston_indep, boston_dep, epochs=1000, verbose=0)
boston_model.fit(boston_indep, boston_dep, epochs=10)
```

```
print(boston_model.predict(boston_indep[:5]))
print(boston_dep[:5])
```

##**Iris**

```
import tensorflow as tf
import pandas as pd
```

```
file_location = 'https://raw.githubusercontent.com/Myeongjung/ML_Practice/master/csv/iris.csv'
iris = pd.read_csv(file_location)
#To make classification, one hot encoding
encoded_data = pd.get_dummies(iris)
print(encoded_data.columns)
encoded_data.head()
```

```
iris_indep = encoded_data[['Petal length', 'Petal width', 'Calyx length', 'Calyx width']]
iris_dep = encoded_data[['Species_setosa', 'Species_versicolor', 'Species_virginica']]

print(iris_indep.shape, iris_dep.shape)
```

```
#The number of independent variable is '4'
X = tf.keras.layers.Input(shape=[4])

layer1 = tf.keras.layers.Dense(8)(X)
layer1 = tf.keras.layers.BatchNormalization()(layer1)
layer1 = tf.keras.layers.Activation('swish')(layer1)

layer2 = tf.keras.layers.Dense(8)(layer1)
layer2 = tf.keras.layers.BatchNormalization()(layer2)
layer2 = tf.keras.layers.Activation('swish')(layer2)

layer3 = tf.keras.layers.Dense(8)(layer2)
layer3 = tf.keras.layers.BatchNormalization()(layer3)
layer3 = tf.keras.layers.Activation('swish')(layer3)

Y = tf.keras.layers.Dense(3, activation='softmax')(layer3)

iris_model = tf.keras.models.Model(X,Y)
iris_model.compile(loss='categorical_crossentropy', metrics='accuracy')
```

```
iris_model.summary()
```

```
iris_model.fit(iris_indep, iris_dep, epochs=1000, verbose=0)
iris_model.fit(iris_indep, iris_dep, epochs=10)
```

```
print(iris_model.predict(iris_indep[:5]))
print(iris_dep[:5])
```
