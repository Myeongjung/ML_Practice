Can't do One-Hot Encoding (get_dummies(iris))

```
#iris2 is the number data
file_location = 'https://raw.githubusercontent.com/Myeongjung/ML_Practice/master/csv/iris2.csv'
iris = pd.read_csv(file_location)
iris.head()
```

```
print(iris.dtypes)
```

Change the data type int -> category

```
iris['Species'] = iris['Species'].astype('category')
print(iris.dtypes)
```

```
Encoded_data = pd.get_dummies(iris)
Encoded_data.head()
```

Fill NaN to Mean value of the Petal width

```
mean = iris['Petal width'].mean()
#print(mean)
iris['Petal width'] = iris['Petal width'].fillna(mean)
iris.tail()
```
